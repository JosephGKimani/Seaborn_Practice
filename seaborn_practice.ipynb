```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")

titanic=sns.load_dataset("titanic")
```

```python
sns.scatterplot(x="tip", y="total_bill",data=tips,hue="day",size="size",palette="YlGnBu")
```

```python
sns.histplot(tips["tip"],kde=True,bins=15)
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette="YlGn",hue="sex")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="total_bill",hue="sex",data=tips,palette="YlGnBu",dodge=True)
```

```python
sns.jointplot(x="day",y="tip",data=tips)
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(),annot=True,cmap="coolwarm")
```
